```python
def extract_count(name):
    if isinstance(name, newick.Node):
        name = name.name
    i = name.rindex('_')
    return int(name[i:][1:])

for node in hog2521.walk(mode="preorder"):
    print(node.name, extract_count(node.name))
```

```python
from itertools import chain
def find_drops(root_node):
    current_count = extract_count(root_node)
    drops = []
    valid_subtrees = list(root_node.descendants)
    for node in root_node.descendants:  # parents first
        if extract_count(node) < current_count:  # found it!
            drops.append(node.name)  # or node)
#             valid_subtrees.remove(node)
        elif extract_count(node) > current_count:
            valid_subtrees.remove(node)
    return list(chain(drops, *[find_drops(t) for t in valid_subtrees]))

find_drops(hog2521)
```

## Global View: Drops

```python
def summarize_drops(tree_str, family_name):
    this_tree = newick.loads(tree_str)[0]
    hits = len(find_drops(this_tree))
    return ('HOG%s' % family_name, hits)
    
summarize_drops(HOG2521, '2521')
```

```python
def summarize(input_path, func):
    data = open(input_path, 'r').read().splitlines()
    drop_counts = []
    for i, line in enumerate(data):
        if i == 0: continue
        parts = line.split('\t')
        name, tree = parts[0], parts[1]
    #     print(tree)
        n, hits = func(tree, name)
        drop_counts.append(hits)
#         print(n, hits)
    return drop_counts
summarize('data/cafe_prototype/reports/run8.csv', summarize_drops)
```

```python
def find_boosts(root_node):
    current_count = extract_count(root_node)
    drops = []
    valid_subtrees = list(root_node.descendants)
    for node in root_node.descendants:  # parents first
        if extract_count(node) > current_count:  # found it!
            drops.append(node.name)  # or node)
#             valid_subtrees.remove(node)
        elif extract_count(node) < current_count:
            valid_subtrees.remove(node)
    return list(chain(drops, *[find_boosts(t) for t in valid_subtrees]))

def summarize_boosts(tree_str, family_name):
    this_tree = newick.loads(tree_str)[0]
    hits = len(find_boosts(this_tree))
    return ('HOG%s' % family_name, hits)

summarize('data/cafe_prototype/reports/run8.csv', summarize_boosts)
```
