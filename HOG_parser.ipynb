```python
def csv_file_into_data_table(count_path):
    with open(count_path, 'r') as count_file:
        r = csv.reader(count_file, delimiter=':')
        frax_hogs = [line for line in r]
        counts = [int(x[1]) for x in frax_hogs]
        names = [x[0].replace('.fa', '') for x in frax_hogs]
    return counts, names

def frax_gene_counts_dataframe(directory):
    import pandas as pd
    import os
    file_list = glob(os.path.join(directory, "*"))
    assert len(file_list), f"Directory '{directory}' is empty or doesn't exist"
    
    all_counts = []
    ignored, names = csv_file_into_data_table(file_list[0])
    for filename in file_list:
        counts, current_names = csv_file_into_data_table(filename)
        assert current_names == names
        all_counts.append(counts)
    
    species_labels = [just_the_name(f).replace('number_of_', '').replace('_sequences_in_each_HOG', '') for f in file_list]
    
    d = {}
    assert len(species_labels) == len(all_counts)
    for i in range(len(species_labels)):
        d[species_labels[i]] = pd.Series(all_counts[i], index=names) 
    data_table = pd.DataFrame(d)
    data_table = data_table.reindex(natsorted(names))
    return data_table
```
