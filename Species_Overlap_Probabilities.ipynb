```python
import pandas
from pandas import read_csv
countpath = r"D:\josiah\Documents\Research\Thesis - Genome Symmetry\DNA_Duplications\CAFE-4.2\data\2020\Feb08_merged_homeolog_counts.csv"
table = read_csv(countpath, delimiter='\t', index_col='Family ID')

table.tail(20)
```

**Note that the table is sorted by total gene count, so the tail has gene families with very few members. This means some amount of gene loss overlap between species is non-random because the families themselves are rare across all species. This can most easily be explained by deep ancestral losses which are not necessarily polyphyletic.***

```python
def geneLossSet(threshold):
    """Gene families below the threshold will be returned."""
    genesLost = dict()
    for species, familyCounts in table.iteritems(): 
        genesLost[species] = set()
        for fam, count in zip(familyCounts.index.values, familyCounts):
            if species == "Oeuropea": 
                count -= threshold # compensate for extra WGD
            if count < threshold:
    #             print(fam)
                genesLost[species].add(fam)
    return genesLost
singleLosses = geneLossSet(2)
doubleLosses = geneLossSet(1)
```

```python
for species, familyCounts in table.iteritems(): 
    print(len(doubleLosses[species]))
```

# Overlap Table

```python
def outputTableFile(losses, tablepath):
    species = table.columns.tolist()
    overlapSingle = []
    for a in species:
        row = []
        for b in species:
            row.append(len(losses[a].intersection(losses[b])))
        overlapSingle.append(row)


    with open(tablepath, 'w') as outfile:
        outfile.write('\t'.join(species) +'\n')
        for row in overlapSingle:
            outfile.write('\t'.join([str(x) for x in row]) +'\n')
            
filepath = r"D:\josiah\Documents\Research\Thesis - Genome Symmetry\DNA_Duplications\CAFE-4.2\data\2020\Feb08_overlap.csv"  
outputTableFile(singleLosses, filepath)
```

```python
filepath = r"D:\josiah\Documents\Research\Thesis - Genome Symmetry\DNA_Duplications\CAFE-4.2\data\2020\Feb08_overlap_two_or_more_losses.csv"  
outputTableFile(doubleLosses, filepath)
```
